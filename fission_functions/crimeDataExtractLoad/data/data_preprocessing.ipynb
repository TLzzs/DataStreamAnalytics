```python
import json

# 1. Read the JSON file
with open('2019_crime_data.json') as f:
    data = f.read()

# 2. Parse the JSON data
json_data = json.loads(data)

```

```python
crime_documents = json_data['features']
```

```python
len(crime_documents)
```

```python
crime_documents[0]['properties']
```

```python
import re

# Define a function to remove bracketed part from a string
def remove_brackets(input_string):
    return re.sub(r'\s*\([^)]*\)', '', input_string)
```

```python
curr = crime_documents[0]['properties']
suburb_name = remove_brackets(curr['lga_name11'])
suburb_name
```

```python
# have a list such that contains the suburb names and the records for crime in that suburb
output_list = []

for i in range(len(crime_documents)):
    curr_suburb_data = crime_documents[i]['properties']
    # Iterate through the dictionary and replace None with 0
    for key, value in curr_suburb_data.items():
        if value is None:
            curr_suburb_data[key] = 0
    # Strip the unwanted bracket info at the end of suburb names
    # Change the name of the key
    curr_suburb_data['suburb_name'] = remove_brackets(curr_suburb_data.pop('lga_name11'))
    output_list.append(curr_suburb_data)
```

```python
output_list
```

```python
# Specify the path to the output JSON file
output_file = '2019_processed_output.json'

# Write the dictionary to the JSON file
with open(output_file, 'w') as f:
    json.dump(output_list, f)
```
